# **プログラムの名称**
ローグライクRPG
## **概要**

RPG形式の戦闘方式のローグライクゲームです。レベルアップなどの方法ではなく、ランダムに選ばれた三つのスキルの中から選択するという形でキャラクターを強化していきます

## **操作方法**

一つ目のコードシェルを実行してから、二つ目のコードシェルを実行することでゲームを開始できます。やり直す際には必ず一つ目のシェルをもう一度実行してください。また、文章を送る際はエンターキーを入力してください。選択肢を選ぶ際にはy/nか1～9までの数字を入力する形で選択してください。戦闘はオートバトル形式です。

## **制作者のコメント**

RPGはコマンド選択でゲームを進める関係上、長い時間を要するゲームですが、本課題においては、短い時間で遊ぶことができるゲームの方が好ましいと思いオートバトル形式で戦闘以外の場面でプレイヤーに選択肢を与えることでゲーム性を与えようとしたのですが、思ったよりもプレイヤーの関与する部分が少なくなってしまったことを少し後悔しています。また、辞書型について学ぶ前に作り始めたためキャラクターのステータスの書き方がすごく効率の悪いものになってしまったのも心残りです。

```python
# @title 初期化＆関数，変数定義
%reset -f

import sys
import time
import random as r
import IPython.display

item_now_buki = []
item_now_bougu = []
At_buki = 0
Df_bougu = 0

#ステータス置き場
#勇者
HP_max_y = 6
HP_y = 6
At_y = 3
Df_y = 3
exp_y = 9999
exp_get = 0
level_y = 0
gold_have = 1000

#テスト戦闘用
name_kakasi = "かかし"
HP_kakasi = 10
At_kakasi = 4
Df_kakasi = 0
exp_kakasi = 5
gold_kakasi = 1


#始まりの平原
name_sougen = "始まりの草原"

#スライム
name_sura = 'スライム'
HP_sura = 8
At_sura = 5
Df_sura = 0
exp_sura = 5
gold_sura = 100

#イノシシ
name_pig =  "フレンジーボア"
HP_pig = 6
At_pig = 7
Df_pig = 2
exp_pig = 8
gold_pig = 200

#ゴーレム
name_golem = "ゴーレム"
HP_golem = 5
At_golem = 4
Df_golem = 7
exp_golem = 10
gold_golem = 200

#なぞに強敵
name_omu = '高専生'
HP_omu = 7
At_omu= 7
Df_omu = 7
exp_omu = 15
gold_omu = 300

#始まりの草原でのエンカウントパターン
enemy_list_sougen_1 = [name_sura,name_pig,name_golem]
enemy_list_sougen_2 = [name_pig,name_golem,name_omu]
enemy_list_sougen_3 = [name_sura]

#初めのボス
name_dog = "ブラッディウルフ"
HP_dog = 20
At_dog = 10
Df_dog = 5
exp_dog = 20
gold_dog = 600

#墓場

name_hakaba = "墓場"

#ゾンビ
name_zonbi = "ゾンビ"
HP_zonbi = 15
At_zonbi = 7
Df_zonbi = 7
exp_zonbi = 0
gold_zonbi = 200

#スケルトン
name_hone = "スケルトン"
HP_hone = 10
At_hone = 9
Df_hone = 5
exp_hone = 0
gold_hone = 200

#幽霊
name_gorst = "幽霊"
HP_gorst = 2
At_gorst = 12
Df_gorst = 0
exp_gorst = 0
gold_gorst = 400

#墓守
name_hakamori = "墓守"
HP_hakamori = 15
At_hakamori = 10
Df_hakamori = 9
exp_hakamori = 0
gold_hakamori = 500

#墓場でのエンカウントパターン
enemy_list_hakaba_1 = [name_zonbi,name_hone,name_gorst]
enemy_list_hakaba_2 = [name_gorst,name_hakamori]

#ケルベロス
name_keruberosu = "ケルベロス"
HP_keruberosu = 40
At_keruberosu = 15
Df_keruberosu = 15
exp_keruberosu = 0
gold_keruberosu = 1500

#魔王城
name_makai = "魔王城"

#デーモン
name_demon = "デーモン"
HP_demon = 30
At_demon = 17
Df_demon = 14
exp_demon = 0
gold_demon = 700

#ガーゴイル
name_gargantua = "ガーゴイル"
HP_gargantua = 15
At_gargantua = 16
Df_gargantua = 20
exp_gargantua = 0
gold_gargantua = 700

#ドラゴン
name_dragon = "ドラゴン"
HP_dragon = 30
At_dragon = 22
Df_dragon = 18
exp_dragon = 500
gold_dragon = 1000

#リッチ
name_rich = "リッチ"
HP_rich = 20
At_rich = 30
Df_rich = 16
exp_rich = 500
gold_rich = 1000

#魔王城でのエンカウントパターン
enemy_list_makai_1 = [name_demon,name_gargantua]
enemy_list_makai_2 = [name_dragon,name_rich]

#魔王
name_maou = "魔王"
HP_maou = 50
At_maou = 30
Df_maou = 25
exp_maou = 1000
gold_maou = 2000

#魔王第二形態
name_maou2 = "魔王"
HP_maou2 = 100
At_maou2 = 35
Df_maou2 = 35
exp_maou2 = 9999
gold_maou2 = 9999
#設定された速度でprint関数を使用する
def print_and_sleep(message):
    print(message)
    time.sleep(let_sp)

#プリント関数とインプット関数を同時に使用する
def print_and_input(message):
    print(message)
    input()

#現在のステータスを表示する関数
def states_open():
    print(f"HP：{HP_max_y}")
    print(f"At：{At_y}")
    print(f"Df：{Df_y}")
    print(f"所持金：{gold_have}")

#戦闘処理
def battle(enemy_name):
    global name_e,HP_e,At_e,Df_e,exp_e,gold_e,HP_max_y,HP_y,At_y,Df_y,exp_get,gold_have
    #敵のステータスをセットする
    if enemy_name == name_kakasi:
        name_e = name_kakasi
        HP_e = HP_kakasi
        At_e= At_kakasi
        Df_e = Df_kakasi
        exp_e = exp_kakasi
        gold_e = gold_kakasi
    elif enemy_name == name_sura:
        name_e = name_sura
        HP_e = HP_sura
        At_e = At_sura
        Df_e = Df_sura
        exp_e = exp_sura
        gold_e = gold_sura
    elif enemy_name == name_pig:
        name_e = name_pig
        HP_e = HP_pig
        At_e = At_pig
        Df_e = Df_pig
        exp_e = exp_pig
        gold_e = gold_pig
    elif enemy_name == name_golem:
        name_e = name_golem
        HP_e = HP_golem
        At_e = At_golem
        Df_e = Df_golem
        exp_e = exp_golem
        gold_e = gold_golem
    elif enemy_name == name_omu:
        name_e = name_omu
        HP_e = HP_omu
        At_e = At_omu
        Df_e = Df_omu
        exp_e = exp_omu
        gold_e = gold_omu
    elif enemy_name == name_dog:
        name_e = name_dog
        HP_e = HP_dog
        At_e = At_dog
        Df_e = Df_dog
        exp_e = exp_dog
    elif enemy_name == name_zonbi:
        name_e = name_zonbi
        HP_e = HP_zonbi
        At_e = At_zonbi
        Df_e = Df_zonbi
        exp_e = exp_zonbi
    elif enemy_name == name_hone:
        name_e = name_hone
        HP_e = HP_hone
        At_e = At_hone
        Df_e = Df_hone
        exp_e = exp_hone
    elif enemy_name == name_gorst:
        name_e = name_gorst
        HP_e = HP_gorst
        At_e = At_gorst
        Df_e = Df_gorst
        exp_e = exp_gorst
    elif enemy_name == name_hakamori:
        name_e = name_hakamori
        HP_e = HP_hakamori
        At_e = At_hakamori
        Df_e = Df_hakamori
    elif enemy_name == name_keruberosu:
        name_e = name_keruberosu
        HP_e = HP_keruberosu
        At_e = At_keruberosu
        Df_e = Df_keruberosu
    elif enemy_name == name_demon:
        name_e = name_demon
        HP_e = HP_demon
        At_e = At_demon
        Df_e = Df_demon
    elif enemy_name == name_gargantua:
        name_e = name_gargantua
        HP_e = HP_gargantua
        At_e = At_gargantua
        Df_e = Df_gargantua
    elif enemy_name == name_dragon:
        name_e = name_dragon
        HP_e = HP_dragon
        At_e = At_dragon
        Df_e = Df_dragon
    elif enemy_name == name_rich:
        name_e = name_rich
        HP_e = HP_rich
        At_e = At_rich
        Df_e = Df_rich
    elif enemy_name == name_maou:
        name_e = name_maou
        HP_e = HP_maou
        At_e = At_maou
        Df_e = Df_maou
    elif enemy_name == name_maou2:
        name_e = name_maou2
        HP_e = HP_maou2
        At_e = At_maou2
        Df_e = Df_maou2
    #print_and_sleep(f"{name_y}の前に{name_e}が現れた")
    print_and_input(f"{name_y}の前に{name_e}が現れた")
    #print_and_sleep(f"{name_y}のHP「{HP_y}」、{name_e}のHP「{HP_e}」")
    print_and_input(f"{name_y}のHP「{HP_y}」、{name_e}のHP「{HP_e}」")
    input()
    IPython.display.clear_output()

    #攻撃力が防御力よりも低いときに、与えるダメージを一に固定する
    if At_e <= Df_y:
        At_e = Df_y + 1
    if At_y <= Df_e:
        Df_e = At_y - 1

    while 1 :
        #主人公の攻撃
        if name_e == name_gorst:
            if r.random() < 0.7:
                print_and_sleep(f"{name_y}の攻撃、幽霊はひらりと身をかわした")
            else:
                if HP_y <= 0:
                    pass
                elif HP_e <= (At_y-Df_e):
                    print_and_sleep(f"{name_y}の攻撃、{name_e}に{(At_y-Df_e)}のダメージ")
                    print_and_sleep(f"{name_e}の HP {HP_e} -> 0")
                else:
                    print_and_sleep(f"{name_y}の攻撃、{name_e}に{(At_y-Df_e)}のダメージ")
                    print_and_sleep(f"{name_e}の HP {HP_e} -> {HP_e - (At_y-Df_e)}")
                HP_e -= At_y-Df_e
        else:
            if HP_y <= 0 :
                pass
            elif HP_e <= (At_y-Df_e):
                print_and_sleep(f"{name_y}の攻撃、{name_e}に{At_y-Df_e}のダメージ")
                print_and_sleep(f"{name_e}の HP {HP_e} -> 0")
            else:
                print_and_sleep(f"{name_y}の攻撃、{name_e}に{At_y-Df_e}のダメージ")
                print_and_sleep(f"{name_e}の HP {HP_e} -> {HP_e - (At_y - Df_e)}")
            HP_e -= At_y - Df_e

        print()

        #エネミーの攻撃

        if HP_e <= 0:
            pass
        elif HP_y <= (At_e-Df_y):
            print_and_sleep(f"{name_e}の攻撃、{name_y}に{(At_e-Df_y)}のダメージ")
            print_and_sleep(f"{name_y}の HP {HP_y} -> 0")
            HP_y -= At_e-Df_y
        else:
            print_and_sleep(f"{name_e}の攻撃、{name_y}に{(At_e-Df_y)}のダメージ")
            print_and_sleep(f"{name_y}の HP {HP_y} -> {HP_y - (At_e-Df_y)}")
            HP_y -= At_e-Df_y

        input()

        print()
        IPython.display.clear_output()

        if HP_e <= 0 or HP_y <= 0:
            break

    if HP_y <= 0:#ゲームオーバー
        for i in range(3):
            print(".",end='')
            time.sleep(1)
        print_and_sleep(f"{name_y}は力尽きた。")
        print_and_sleep("ゲームを終了したください。")
        sys.exit(0)

    print_and_sleep(f"{name_y}は{name_e}との戦闘に勝利した!!")
    #time.sleep(1)
    input()
    IPython.display.clear_output()

    #goldの獲得
    print_and_sleep(f"{gold_e}のゴールドを獲得した")
    gold_have += gold_e
    print_and_sleep(f"現在の所持金が{gold_have}になった")
    #time.sleep(1)
    input()
    IPython.display.clear_output()

    '''#レベルアップという概念を無くしたのでいらなくなった
    #経験値の処理
    print_and_sleep(f"{name_y}は{exp_e}の経験値を手に入れた")
    exp_get += exp_e
    print_and_sleep(f"累計経験値 -> {exp_get}")

    if exp_get >= 10 and level_y <= 1:
        level_up(2)
        print_and_sleep(f"次のレベルアップまでは経験値{20 - exp_get}")
    elif exp_get >= 20 and level_y <= 2:
        level_up(3)
        print_and_sleep(f"次のレベルアップまでは経験値{40 - exp_get}")
    elif exp_get >= 40 and level_y <= 3:
        level_up(4)
        print_and_sleep(f"次のレベルアップまでは経験値{70 - exp_get}")
    elif exp_get >= 70 and level_y <= 4:
        level_up(5)
        print_and_sleep(f"次のレベルアップまでは経験値{110 - exp_get}")
    elif exp_get >= 110 and level_y <= 5:
        level_up(6)
        print_and_sleep(f"次のレベルアップまでは経験値{160 - exp_get}")
    elif exp_get >= 160 and level_y <= 6:
        level_up(7)
        print_and_sleep(f"次のレベルアップまでは経験値{230 - exp_get}")
    elif exp_get >= 230 and level_y <= 7:
        level_up(8)
        print_and_sleep(f"次のレベルアップまでは経験値{350 - exp_get}")
    elif exp_get >= 350 and level_y <= 8:
        level_up(9)
        print_and_sleep(f"次のレベルアップまでは経験値{500 - exp_get}")
    elif exp_get >= 500 and level_y <= 9:
        level_up(10)
        print_and_sleep(f"次のレベルアップまでは経験値{700 - exp_get}")
    '''
    IPython.display.clear_output()

#勇者探索中...と出力する
def loading():
    print("勇者探索中", end="")
    time.sleep(0.5)
    for i in range(3):
        print(".", end="")
        time.sleep(0.5)

#エンカウント処理
def encount(enemy_list):
    enemy_name = r.choice(enemy_list)
    battle(enemy_name)


#エンカウントなどのパターンを設定し戦闘などを起こす
def field(field_name,enemy_list_1,encount_count_1,enemy_list_2,encount_count_2,enemy_list_3,encount_count_3,skil_list):
    print_and_input(f"{name_y}は{field_name}に足を踏み入れた")
    for i in range(encount_count_1):
        print_and_sleep(f"{name_y}は{field_name}を進む")
        loading()
        encount(enemy_list_1)
        skil(skil_list)
        print_and_sleep(f"薬草を使用しますか? [y/n]")
        player_select_item = input()
        if player_select_item == "y":
            yakusou_use()
    for i in range(encount_count_2):
        print_and_sleep(f"{name_y}は{field_name}を進む")
        loading()
        encount(enemy_list_2)
        skil(skil_list)
        print_and_sleep(f"薬草を使用しますか? [y/n]")
        player_select_item = input()
        if player_select_item == "y":
            yakusou_use()
    for i in range(encount_count_3):
        print_and_sleep(f"{name_y}は{field_name}を進む")
        loading()
        encount(enemy_list_3)
        skil(skil_list)
        print_and_sleep(f"薬草を使用しますか? [y/n]")
        player_select_item = input()
        if player_select_item == "y":
            yakusou_use()

def add_at_buki(buki_name):
    global At_y,At_buki
    if buki_name in item_buki:
        At_buki = item_buki[buki_name]
        At_y += At_buki

def add_df_bougu(bougu_name):
    global Df_y,Df_bougu
    if bougu_name in item_bougu:
        Df_bougu = item_bougu[bougu_name]
        Df_y += Df_bougu

def item_get(item_name): #装備を獲得する関数
    global item_now_buki, item_now_bougu, At_y, Df_y, At_buki, Df_bougu
    if item_name in item_buki:
        if len(item_now_buki) == 0:  #現在の武器が空の時には装備をそのまま獲得する
            item_now_buki.append(item_name)
            add_at_buki(item_name)
            print_and_sleep(f"{item_name}を手に入れた")
            input()
            IPython.display.clear_output()
        else: #現在装備を持っていれば獲得するかを確認する
            print_and_sleep(f"既に装備を持っています")
            print_and_sleep(f"{item_name}と入れ替えますか？ [y/n]")
            player_select_item = input()
            if player_select_item == "y":# 現在の武器の攻撃力を減算
                At_y -= item_buki[item_now_buki[0]]# 新しい武器を装備
                item_now_buki[0] = item_name
                add_at_buki(item_name)
                print_and_sleep(f"{item_name}を手に入れた")
            elif player_select_item == "n":
                print_and_sleep(f"{item_name}を捨てた")
            input()
            IPython.display.clear_output()
    elif item_name in item_bougu:
        if len(item_now_bougu) == 0:#現在の防具が空の時には装備をそのまま獲得する
            item_now_bougu.append(item_name)
            add_df_bougu(item_name)
            print_and_sleep(f"{item_name}を手に入れた")
        else: #現在装備を持っていれば獲得するかを確認する
            print_and_sleep(f"既に装備を持っています")
            print_and_sleep(f"{item_name}と入れ替えますか？ [y/n]")
            player_select_item = input()
            if player_select_item == "y": # 現在の防具の防御力を減算
                Df_y -= item_bougu[item_now_bougu[0]]# 新しい防具を装備
                item_now_bougu[0] = item_name
                add_df_bougu(item_name)
                print_and_sleep(f"{item_name}を手に入れた")
            elif player_select_item == "n":
                print_and_sleep(f"{item_name}を捨てた")

item_buki = {
        "きのぼう":4,
        "てつのつるぎ":9,
        "はがねのつるぎ":15,
        "エクスカリバー":30
}

item_bougu = {"ぬのきれ":3,
              "てつのよろい":8,
              "はがねのよろい":13
}

# ステータスをアップデートする関数
def states_up(state_name, amount_states_up):
    global HP_max_y, At_y, Df_y, exp_get, gold_have
    if state_name == "HP_max_y":
        HP_max_y += amount_states_up
    elif state_name == "At_y":
        At_y += amount_states_up
    elif state_name == "Df_y":
        Df_y += amount_states_up
    elif state_name == "exp_get":
        exp_get += amount_states_up
    elif state_name == "gold_have":
        gold_have += amount_states_up

#スキルを取得する
def skil(skil_list):
    selected_skil = r.sample(list(skil_list.keys()), 3)
    print("取得するスキルを選択してください")
    for i, key in enumerate(selected_skil, start=1):
        print(f"{i}: {key}")
    choice = int(input("取得するスキルの番号を入力してください")) - 1
    if 0 <= choice < len(selected_skil):
        skil_list[selected_skil[choice]]()
    else:
        print("Invalid choice")
    states_open()
    input()
    IPython.display.clear_output()

# スキルリスト
skil_dict_sougen = {
    "At_up": lambda: states_up("At_y", 2),
    "Df_up": lambda: states_up("Df_y", 2),
    "HP_up": lambda: states_up("HP_max_y", 4),
    "gold_get": lambda: states_up("gold_have", 200),
    "きのぼう": lambda: item_get("きのぼう"),
    "ぬのきれ": lambda: item_get("ぬのきれ"),
    "full_HP": lambda: recovery()
}
skil_dict_hakaba = {
    "At_up": lambda: states_up("At_y", 2),
    "Df_up": lambda: states_up("Df_y", 2),
    "HP_up": lambda: states_up("HP_max_y", 4),
    "gold_get": lambda: states_up("gold_have", 400),
    "てつのつるぎ": lambda: item_get("てつのつるぎ"),
    "てつのよろい": lambda: item_get("てつのよろい"),
    "full_HP": lambda: recovery()
}
skil_dict_makai = {
    "At_up": lambda: states_up("At_y", 2),
    "Df_up": lambda: states_up("Df_y", 2),
    "HP_up": lambda: states_up("HP_max_y", 4),
    "gold_get": lambda: states_up("gold_have", 600),
    "はがねのつるぎ": lambda: item_get("はがねのつるぎ"),
    "はがねのよろい": lambda: item_get("はがねのよろい"),
    "full_HP": lambda: recovery()
}

#薬草関連
def recovery():
    global HP_y
    print_and_sleep(f"HPが回復した  : {HP_y} -> {HP_max_y}")
    HP_y = HP_max_y
    input()
    IPython.display.clear_output()

yakusou = {"やくそう"  :2}

def yakusou_use():
    global yakusou
    if yakusou['やくそう'] > 0:
        print("やくそうを使用した")
        recovery()
        yakusou_lost()
    else:
        print("やくそうがありません")

def yakusou_get():
    global yakusou
    yakusou['やくそう']+=1
    print_and_sleep(f"薬草を獲得した")
    print_and_input(f"やくそう：{yakusou['やくそう']}")

def yakusou_lost():
    global yakusou
    if yakusou['やくそう'] > 0:
      yakusou['やくそう']-=1
      print(f"やくそう：{yakusou['やくそう']}")

def city(yadoya_price,yakusou_price,kunnren_price,):
    global gold_have
    print_and_sleep(f"{name_y}は町にたどり着いた")
    print_and_sleep(f"何か購入しますか？")
    print("1 : 宿屋")
    print("2 : アイテム屋")
    print("3 : 訓練場")
    print("4 : 旅に出る")
    player_select_city = input()
    if player_select_city == "1":
        print_and_sleep(f"{name_y}は宿屋に入った")
        print_and_sleep(f"現在の所持金{gold_have}")
        print_and_sleep(f"一泊{yadoya_price}Gです。宿泊しますか？[y/n]")
        player_select_yadoya = input()
        if player_select_yadoya == "y":
            print_and_sleep(f"HPが全回復した")
            recovery()
            gold_have -= yadoya_price
            print_and_sleep(f"現在の所持金が{gold_have}になった")
            input()
            IPython.display.clear_output()
            city(yadoya_price,yakusou_price,kunnren_price,)
        else :
            print_and_sleep("宿屋を出た")
            input()
            IPython.display.clear_output()
            city(yadoya_price,yakusou_price,kunnren_price,)
    elif player_select_city == "2":
         print_and_sleep(f"{name_y}はアイテム屋に入った")
         print_and_sleep(f"現在の所持金{gold_have}")
         print_and_sleep("薬草が店頭に並んでいます")
         print_and_sleep(f"薬草の値段{yakusou_price}Gです。薬草を購入しますか？[y/n]")
         player_select_yakusou = input()
         if player_select_yakusou == "y":
            yakusou_get()
            gold_have -= yakusou_price
            print_and_sleep(f"現在の所持金が{gold_have}になった")
            input()
            IPython.display.clear_output()
            city(yadoya_price,yakusou_price,kunnren_price,)
         else :
            print_and_sleep("アイテム屋を出た")
            input()
            IPython.display.clear_output()
            city(yadoya_price,yakusou_price,kunnren_price,)
    elif player_select_city == "3":
         print_and_sleep(f"{name_y}は訓練場に入った")
         print_and_sleep(f"現在の所持金{gold_have}")
         states_open()
         print_and_sleep(f"訓練の値段は{kunnren_price}Gです。訓練を開始しますか？[y/n]")
         player_select_kunnren = input()
         if player_select_kunnren == "y":
            print_and_sleep(f"訓練するステータスを選択してください")
            print("1 : HP")
            print("2 : At")
            print("3 : Df")
            player_select_states = input()
            gold_have -= kunnren_price
            print_and_sleep(f"現在の所持金が{gold_have}になった")
            if player_select_states == "1":
               states_up("HP_max_y", 4)
               print_and_sleep(f"現在のHPが{HP_max_y}になった")
               states_open()
            elif player_select_states == "2":
               states_up("At_y", 2)
               print_and_sleep(f"現在のAtが{At_y}になった")
               states_open()
            elif player_select_states == "3":
               states_up("Df_y", 2)
               print_and_sleep(f"現在のDfが{Df_y}になった")
               states_open
            input()
            IPython.display.clear_output()
            city(yadoya_price,yakusou_price,kunnren_price,)
         else :
          print_and_sleep("訓練場を出た")
          input()
          IPython.display.clear_output()
          city(yadoya_price,yakusou_price,kunnren_price,)
    elif player_select_city == "4":
         print_and_sleep("町を出て旅に出た")
         input()
         IPython.display.clear_output()
```
